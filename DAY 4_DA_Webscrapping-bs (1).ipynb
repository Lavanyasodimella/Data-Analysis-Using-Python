Extarcting HTML code of a webpage

program flow
1. Import libraries
2. save the url
3. Using request.get(),access the web page
4. Using BeautifulSoup(),access the HTML code

```python
from bs4 import BeautifulSoup as bs
import requests as rq
```

```python
url ="https://www.netflix.com/in/"
```

```python
page=rq.get(url)
```

```python
code=bs(page.text,"html")
print(code)
```

```python
t= code.find("title")
print(t)
```

```python
print(t.text)
```

Exracting a HTML from a webpage

```python
rich="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
```

```python
ipl=rq.get(rich)
```

```python
ppl=bs(ipl.text,"html")
print(ppl)
```

```python
ccl=ppl.find("table")
print(ccl)
```

```python
ccl1=ccl.find_all("th")
print(ccl1)
```

```python
heading=[(i.text) for i in ccl1]
print(heading)
```

```python
import pandas as pd
```

```python
mydf=pd.DataFrame(columns=heading)
print(mydf)
```

```python
rows=ccl.find_all("tr")
print(rows)
```

```python
td=ccl.find_all("td")
print(td)
```

```python
row=[(i.text) for i in td]
print(row)
```

```python
print(row[0:4])
```

```python
st=0
stop=4
ind=0
for i in row:
  if stop<=40:
    mydf.loc[ind]=(row[st:stop])
    st+=4
    stop+=4
    ind=ind+1
print(mydf)
```
