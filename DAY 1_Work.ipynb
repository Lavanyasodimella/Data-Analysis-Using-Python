```python
#create a set
s1={1,55,34,45,34,90,67,66} #Duplicate values are not allowed
print(s1)
```

```python
#sets are unchangeable we cananot change the elements in set but we can add elements into a set
s1.add(110)
print(s1)
```

```python
#add one set of elements to another set
s2={45,932875,7687,10,5758}
s1.update(s2)
print(s1)
```

```python
#remove elements from set
s3={22,11,66,44,99}
s3.remove(99)
print(s3)
```
