```python
#there are five subjects psychology,nanotechnology,medicine,oncology,anthropology two students have applied todo masters in any of these five fileds with a line plot show in which field they can excel . the cutoff for every sub is 70
sub=["psychology","nano technology","medicine"," onocology","anthrapology"]
s1=[67,45,73,71,57]
s2=[43,67,68,53,85]
cut=[70,70,70,70,70]
plt.plot(sub,s1,label="Student1",color="pink",marker="p")      #use scatter instead of plot ,we get dots instead of lines
plt.plot(sub,s2,label="Student2",color="blue",marker="H")
plt.plot(sub,cut,label="Cutoff",color='m',linestyle=":",linewidth=3)
plt.legend()
plt.xlabel("Subjects")
plt.ylabel("Marks")
```

# ***#take overs from arange fun create runs as a array,linewidth is 2:\***


```python
import numpy as np
overs1=np.arange(5,51,5)
runind=np.array([25,51,84,131,160,189,220,250,267,297])
overs2=np.arange(5,26,5)
runaus=np.array([15,41,94,110,151])
plt.plot(overs1,runind,label="India",color="blue",marker='>',linewidth=2)
plt.plot(overs2,runaus,label="Australia",color="yellow",marker='P',linewidth=2)
plt.title("Ind vs Aus")
plt.xlabel("overs")
plt.ylabel("runs")
plt.legend()
plt.grid()
```
