```python
#hypothesis testing
#import
import numpy as np
from scipy import stats
```

```python
# two groups of data
data1 = np.array([28,21,26,29,23])
data2 = np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
# perform t-test assuming unequal variances
t_stat , p_value = stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#print the results
print("T - statistic:",t_stat)
print("P - value:",p_value)
```

```python
# set significance level
alpha = 0.05
```

```python
# interupt the results
if p_value <alpha :
  print("reject the null hypothesis.the means are significant")
else :
  print("fail to reject the null hypothesis. the means are not significant")
```

```python
import pandas as pd
df = pd.read_csv("/content/Vaccine.csv")
```

```python
df
```

```python
a=df["Efficiency"].mean()
print("mean efficiency",a)
b=df["Rec_Rate"].mean()
print("mean rc_rate",b)
data1=df["Efficiency"]
data2=df['Rec_Rate']
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
print(t_stat)
print(p_value)
```

```python
alpha = 0.05
```

```python
if p_value > alpha:
  print("H0 - related")
else:
  print("H1 - not related/significantly different")
```

ANOVA TESTING(ONE WAY ANOVA) is for f disbrution s for hypothesis

```python
dfa = df[df['Vaccine']=='A']['Efficiency']
dfb =  df[df['Vaccine']=='B']['Efficiency']
dfa
```

```python
f_stat,p_value=stats.f_oneway(dfa,dfb)

print("F-statistic:",f_stat)
print("p-value:",p_value)
```

```python
if p_value > alpha:
  print("H0 - related")
else:
  print("H1 - not related/significantly different")
```
