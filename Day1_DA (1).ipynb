```python
#take a number from the user which is a salary ,if the salary is >50000 there is a 10% tax print his final salary
salary=int(input("Enter salary:"))
if(salary>=50000):
  salary=salary-(salary*0.1)
  print("final salry is:",salary)
else:
  print("your salary is less than 50000 so no tax  ")
```

```python
#get the salary from the user if the salary is more than or equal 50000 apply 10% tax deduction , if the salary is lessthan 50000 apply 5% tax,if the salary is more than 70000 apply 12% tax,if the salary exceeds 1L you are promoted to next level
salary=int(input("Enter salary"))
if(salary==0):
  print("give valid salary")
elif(salary<=50000):
  salary=salary-(salary*(5/100))
  print("your salary is:",salary)
elif(salary<=70000):
  salary=salary-(salary*(10/100))
  print("your salary is:",salary)
elif(salary<=100000):
  salary=salary-(salary*(12/100))
  print("your salary is:",salary)
else:
  print("your promoted to next level")
```
