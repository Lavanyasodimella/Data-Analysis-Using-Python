```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as rq
```

```python
building=("https://en.m.wikipedia.org/wiki/List_of_tallest_buildings_in_India")
```

```python
page=rq.get(building)
```

```python
code=bs(page.text,'html')
print(code)
```

```python
t1=code.find("table")
print(t1)
```

```python
t2=t1.find_all("th")
print(t2)
```

```python
heading=[(i.text) for i in t2 ]
print(heading)
```

```python
import pandas as pd
```

```python
dataframe=pd.DataFrame(columns=heading)
print(dataframe)
```

```python
t3=code.find_all("tr")
print(t3)
```

```python
t4=code.find_all("td")
print(t4)
```

```python
row=[(i.text) for i in t4]
print(row)
```

```python
print(row[0:9])
```

```python
start=0
stop=9
ind=1
for i in row:
  if stop<=692:
    dataframe.loc[ind]=(row[start:stop])
    start=start+9
    stop=stop+9
    ind=ind+1
print(dataframe)
```
